Pre Pruning - 
limiting the size of a tree during the recursive partitioning process 

This is efficient but may result in important data loss

Post Proning, limit the size of the tree after the partioning.

This does not lose data - but its not efficient 

Post Pruning - Cost Complexity or weakest link

```python
path = regressor.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas
list(ccp_alphas)
```

```python
ccp_alphas = ccp_alphas[:-1]
```

```python
train_scores, test_scores = [], []
for alpha in ccp_alphas:
    regressor_ = DecisionTreeRegressor(random_state=1234, ccp_alpha=alpha)
    model__ = regressor_.fit(X_train, y_train)
    train_scores.append(model__.score(X_train, y_train))
    test_scores.append(model__.score(X_test, y_test))
```

```python
train_scores
```

```python
train_scores
```

```python
plt.plot(ccp_alphas, train_scores, marker = 'o', label='train_score', drawstyle='steps-post')
plt.plot(ccp_alphas, test_scores, marker = 'o', label='test_score', drawstyle='steps-post')
plt.legend()
plt.title('R-Squared by alpha')
```

```python
ix = test_scores.index(max(test_scores))
best_alpha = ccp_alphas[ix]
best_alpha
```

```python
regressor_ = DecisionTreeRegressor(random_state=1234, max_depth=10, min_samples_leaf=2,min_samples_split=2, ccp_alpha=best_alpha)
model_ = regressor_.fit(X_train, y_train)
```

```python
model_.score(X_train, y_train)
```

```python
model_.score(X_test, y_test)
```

```python
plot_tree(
    model_, 
    feature_names = X_train.columns,
    filled = True
)
```
