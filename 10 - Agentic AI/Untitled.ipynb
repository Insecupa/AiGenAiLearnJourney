```python
from langchain.tools import tool
```

```python
@tool("Squire_root")
def square_root(x: float) -> float:
    """Calculaaaate the square root of a number"""
    return x + 0.5
```

```python
square_root.invoke({"x" : 45454})
```

```python
math_agent = create_agent(
    model=gpt4o_chat,
    tools = [square_root],
    system_prompt='You are a math wizard, you can calcualte a root of any number'
)
```

```python
questions = HumanMessage(content="What is the ssquare root of 45454")

response = math_agent.invoke({"messages" : [questions]})

print(response['messages'][-1].content)
```

```python
from typing import Dict, Any
from tavily import TavilyClient

tavily_client = TavilyClient()

@tool
def web_search(query: str) -> Dict[str, Any]:
    """Search the web for information"""
    
    return tavily_client.search(query)

web_search.invoke("Whos the current MLA of Hyderabad area")
```

```python
llm_agent = create_agent(
    model=gpt4o_chat,
    tools=[web_search]
)

questions = HumanMessage(content="Whos the current MLA of Hyderabad area")

response = llm_agent.invoke(
    {"messages" : [questions]}
)
```

```python
response['messages']
```

### Memory

```python
from langgraph.checkpoint.memory import InMemorySaver

agent = create_agent(
    model = gpt4o_chat,
    checkpointer = InMemorySaver()
)
```

```python
question = HumanMessage(content="MY name is jittoBoy and i havent decided my fav color")

config = {'configurable' : {"thread_id" : '1'}}

response = agent.invoke(
    {"messages" : [question]},
    config
)
```
